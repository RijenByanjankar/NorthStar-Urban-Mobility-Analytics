```python
# Example MongoDB-style indexes for optimisation

indexes = {
    "customer_cases": [
        "customer_id",
        "order_id",
        "complaint.status",
        "complaint.severity"
    ],
    "deliveries": [
        "delivery_status",
        "pickup_zone",
        "driver_id",
        "vehicle_id"
    ],
    "app_events": [
        "customer_id",
        "order_id",
        "event_type",
        "success_flag"
    ]
}

indexes
```

## Optimisation Explanation

Indexes were proposed on frequently searched fields such as customer_id, order_id, delivery_status, pickup_zone, vehicle_id, and complaint severity. These indexes improve query performance because MongoDB can locate relevant records faster instead of scanning every document.

For customer case analysis, indexes on customer_id and order_id support faster retrieval of complaint histories and app events. Indexes on complaint.status and complaint.severity help managers quickly identify unresolved or high-risk cases.

For delivery analysis, indexes on delivery_status and pickup_zone help identify failed or delayed deliveries by location. Indexes on driver_id and vehicle_id support performance analysis of drivers and maintenance-related issues.

For app event analysis, indexes on event_type and success_flag help detect platform failures such as unsuccessful chat escalation or payment retry events.

The document-based MongoDB design also reduces the need for repeated joins because related complaint and app event information can be stored together in nested documents.
