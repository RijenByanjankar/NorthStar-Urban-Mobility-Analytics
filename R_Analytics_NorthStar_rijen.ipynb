```r
deliveries <- read_csv("deliveries.csv")

head(deliveries)
```

```r
deliveries %>%
  group_by(delivery_status) %>%
  summarise(avg_cost = mean(fuel_or_charge_cost))
```

```r
ggplot(deliveries, aes(x = delivery_status)) +
  geom_bar() +
  ggtitle("Delivery Status Distribution")
```

```r
deliveries %>%
  group_by(delivery_status) %>%
  summarise(avg_rating = mean(customer_rating_post_delivery, na.rm = TRUE))
```
